- Crie um conjunto


```python
conjunto = {"var", "legal", "par"}
```

- Converta uma lista em um conjunto


```python
lista = [1,2,3]
conjunto_lista = set(lista)
print(conjunto_lista)
```

- Insira novos valores no conjunto


```python
conjunto_lista.add(4)
```

- Remova um elemento do conjunto


```python
conjunto_lista.remove(1) #gera erro caso não tenha valor
conjunto_lista.discard(1) #não gera erro caso não tenha o valor
print(conjunto_lista)
```

- Verifique se um elemento está contino no conjunto

```python
if 2 in conjunto_lista:
    print("legal")
```
