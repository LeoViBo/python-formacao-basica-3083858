- Crie uma classe com 3 atributos

- Defina um método para a classe

- Instancie a classe

- Execute o método

```python
class classe_legal():
    def __init__(self, var1, var2, var3):
        self.var1 = var1
        self.var2 = var2
        self.var3 = var3
    
    def print_vars(self):
        print(f"{self.var1}\n{self.var2}\n{self.var3}")
```

```python
val_teste = classe_legal(1, "texto", True)
val_teste.print_vars()
```
