- Crie tratamento para as exceções: ZeroDivisionError e TypeError usando a sintaxe try...except...finally


```python
try:
    i = 2/0
except ZeroDivisionError:
    print("Não da para dividir por 0")
except TypeError:
    print("Erro de tipagem")
except Exception:
    print("Erro")
finally:
    print("Fim")
    
```
