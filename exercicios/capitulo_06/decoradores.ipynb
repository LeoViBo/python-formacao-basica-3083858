- Crie um exemplo de função e em seguida crie um decorador para alterar o seu comportamento

```python
def decorador_legal(func):
    def coisa_legal(*args):
        print("que legal")
        funcoes = func(*args)
        return funcoes
    return coisa_legal

@decorador_legal
def outra_funcao():
    print("par é legal")

outra_funcao()
```
