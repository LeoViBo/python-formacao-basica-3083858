- Crie uma classe com 3 atributos


- Defina uma variável de classe


- Modifique a variável de instância "cor"


- Printe na tela a variável de classe "cor"


- Crie uma variável privada


```python
class gato():
    _cor="preta"
    def __init__(self, nome, tamanho):
        self.nome = nome
        self.tamanho = tamanho

    def setCor(self, cor):
        self._cor = cor
    
    def getCor(self):
        return self._cor

gato_legal = gato("mia", "pequena")
gato_legal.setCor("Tigrada")
print(gato_legal.getCor())
```
