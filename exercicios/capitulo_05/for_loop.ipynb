- Crie uma lista


```python
lista = [1,2,3,4,5]
```

- Crie um for loop para imprimir na tela todos os elementos da lista


```python
for i in lista:
    print(i)
```

- Crie um for loop usando o recurso "continue"


```python
for i in lista:
    if i%2 == 0:
        continue
    print(i)
```

- Crie um for loop usando o recurso "break"


```python
for i in lista:
    if i%2 == 0:
        continue
    if i >= 5:
        break
    print(i)
```

- Crie um for loop usando o "else"


```python
for i in lista:
    if i%2 == 0:
        continue
    if i >= 5:
        break
    print(i)
else:
    print("rodou")
```
